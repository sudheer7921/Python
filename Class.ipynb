```python
class Add: #클래스는 콜하면 인스턴스 생성
    def __init__(self,x,y):
        self.x = x
        self.y = y
        
    def __call__(self):
        return self.x + self.y 
    
a = Add(5,5) # a 라는 객체 생성(?)
print( a() )
print( a.__call__ )
#print( dir(sum) )
print(a.__dict__)
```
